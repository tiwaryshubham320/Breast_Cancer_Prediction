```python
df['diagnosis']=df['diagnosis'].map({'M':1, 'B':0})
df.head()
```

```python
df['diagnosis'].value_counts()
```

```python
X=df.drop(['diagnosis'], axis=1)
X
```

```python
y=df['diagnosis']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.transform(X_test)
```

```python
X_train
```

```python
X_train.shape
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_pred=lr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
lr_acc=accuracy_score(y_test, y_pred)
lr_acc
```

```python
results=pd.DataFrame()
```

```python
temp_results=pd.DataFrame({'Algorithm':'Linear Regression method', 'Accuracy':[lr_acc]})
results=pd.concat([results, temp_results])
results=results[['Algorithm', 'Accuracy']]
results
```

```python
#Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(X_train, y_train)
```

```python
y_pred=dtc.predict(X_test)
y_pred
```

```python
dtc_acc=accuracy_score(y_test, y_pred)
dtc_acc
```

```python
temp_results=pd.DataFrame({'Algorithm':'Decision Tree Classifier', 'Accuracy':[dtc_acc]})
results=pd.concat([results, temp_results])
results=results[['Algorithm', 'Accuracy']]
results
```

```python
#Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(X_train, y_train)
```

```python
y_pred=rfc.predict(X_test)
y_pred
```

```python
rfc_acc=accuracy_score(y_test, y_pred)
rfc_acc
```

```python
temp_results=pd.DataFrame({'Algorithm':'Random Forest Classifier', 'Accuracy':[rfc_acc]})
results=pd.concat([results, temp_results])
results=results[['Algorithm', 'Accuracy']]
results
```

```python
#Support Vector classifier (SVC)
from sklearn.svm import SVC
svc=SVC()
svc.fit(X_train, y_train)
```

```python
y_pred=svc.predict(X_test)
y_pred
```

```python
svc_acc=accuracy_score(y_test, y_pred)
svc_acc
```

```python
temp_results=pd.DataFrame({'Algorithm':'Random Forest Classifier', 'Accuracy':[rfc_acc]})
results=pd.concat([results, temp_results])
results=results[['Algorithm', 'Accuracy']]
results
```
